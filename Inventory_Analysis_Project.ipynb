# Inventory Reorder Point Analysis
This project calculates when to order warehouse stock using Python and export a dashboard ready file to Excel.

```python
# --- STEP 1: IMPORT TOOLS ---
# Pandas is like Excel for Pyhton. It handles tables.
import pandas as pd
import numpy as np
from google.colab import files

# --- STEP 2: CREATE THE DATA ---
# This ia our "Mock" warehouse data
# In Python, we call this a 'Dictionary'
data = {
    'Product_Name': ['Cordless Drill', 'Hammer', 'Paint', 'Lumber', 'Glue'],
    'Current_Stock': [45, 110, 12, 200, 30],
    'Daily_Sales': [5, 12, 8, 25, 4],
    'Lead_Time_Days': [7, 3, 5, 10, 4]
}

# Turn the data into a Table (DataFrame)
df = pd.DataFrame(data)

# --- STEP 3: THE CALCULATION ---
# We want a 10 unit "Safety Buffer"
# Reorder Point = (Sales per day * Days to wait) + Buffer
df['Reorder_Point'] = (df['Daily_Sales'] * df['Lead_Time_Days']) + 10

# --- STEP 4: CREATE THE ALERT ---
# Tell us if we need to buy more
# If Current_Stock is less than or equal to the Reorder_Point, mark as 'REORDER'
df['Status'] = np.where(df['Current_Stock'] <= df['Reorder_Point'], 'REORDER', 'OK')

# --- STEP 5: SHOW THE RESULT IN COLAB
print("--- Warehouse Inventory Report ---")
print(df)

# Save to Excel
df.to_excel('inventory_analysis.xlsx', index=False)

# This line pops up the download window on your computer
files.download('inventory_analysis.xlsx')
```
